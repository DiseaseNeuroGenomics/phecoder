### Compute and save embeddings for a range of LLMs (run this section with GPU)

```python
embedding_dir = os.path.join(main_dir, 'data/embeddings')
models = [
    "Alibaba-NLP/gte-Qwen2-7B-instruct"
    # "all-MiniLM-L6-v2",
    # "NeuML/pubmedbert-base-embeddings",
    # "FremyCompany/BioLORD-2023",
    # "cambridgeltl/SapBERT-from-PubMedBERT-fulltext-mean-token",
]
```

```python
for model_name in models:
    icd_df_temp = icd_df.copy()
    phecode_df_temp = phecode_df.copy()
    model = SentenceTransformer(model_name, device="cpu")  

    icd_embeddings = model.encode(icd_df["ICDString"].tolist(),
                            show_progress_bar=True,
                            normalize_embeddings=True,
                            batch_size=64)  
    icd_df_temp["icd_embedding"] = list(icd_embeddings)

    phecode_embeddings = model.encode(phecode_df["PhecodeString"].tolist(),
                            show_progress_bar=True,
                            normalize_embeddings=True,
                            batch_size=64)  
    phecode_df_temp["phecode_embedding"] = list(phecode_embeddings)

    save_path = os.path.join(embedding_dir, model_name.split('/')[-1]) 
    os.makedirs(save_path, exist_ok=True)
    icd_df_temp.to_parquet(os.path.join(save_path, 'icd_embeddings.parquet'))
    phecode_df_temp.to_parquet(os.path.join(save_path, 'phecode_embeddings.parquet'))

    
```

### Testing

```python
from sklearn.metrics import precision_recall_curve, average_precision_score

model_metrics = {}

for model_name in models:

    icd_df = pd.read_parquet(os.path.join(embedding_dir, model_name.split('/')[-1], 'icd_embeddings.parquet'))
    phecode_df = pd.read_parquet(os.path.join(embedding_dir, model_name.split('/')[-1], 'phecode_embeddings.parquet'))
    phecodes = phecode_df['Phecode'].dropna().unique().tolist()

    all_curves = []
    y_pool, s_pool = [] , []

    for phecode in phecodes[:100]:

        phe_embedding = phecode_df[phecode_df['Phecode']==phecode]['phecode_embedding'].to_numpy()[0]

        mask = icd_df["icd_embedding"].notna()

        E = np.vstack(icd_df.loc[mask, "icd_embedding"].to_numpy())  # shape (n, d)
        sims = E @ phe_embedding  # dot == cosine for unit vectors

        icd_sorted = (
            icd_df.loc[mask]
                .assign(cosine_sim=sims)
                .sort_values("cosine_sim", ascending=False)
        )

        icd_sorted['label'] = 0
        icd_sorted.loc[icd_sorted['Phecode']==phecode, 'label'] = 1

        y_true = np.asarray(icd_sorted['label'], dtype=int) 
        scores = np.arange(len(y_true), 0, -1)
        P, R, _ = precision_recall_curve(y_true, scores)

        all_curves.append((R, P))

        # add to micro pools
        y_pool.append(y_true)
        s_pool.append(scores)

    # ---------- micro‑average PR curve ------------------------------
    y_all = np.concatenate(y_pool)
    s_all = np.concatenate(s_pool)

    P_micro, R_micro, _ = precision_recall_curve(y_all, s_all)
    ap_micro = average_precision_score(y_all, s_all)

    model_metrics[model_name] = (P_micro, R_micro, ap_micro)
```

```python
from sklearn.metrics import precision_recall_curve, average_precision_score
from joblib import Parallel, delayed

model_metrics = {}

for model_name in models:

    path = os.path.join(embedding_dir, model_name.split('/')[-1])
    icd_df = pd.read_parquet(os.path.join(path, 'icd_embeddings.parquet'))
    phecode_df = pd.read_parquet(os.path.join(path, 'phecode_embeddings.parquet'))
    phecodes = phecode_df['Phecode'].dropna().unique().tolist()

    # Precompute once for all phecodes
    mask = icd_df["icd_embedding"].notna().to_numpy()
    E = np.vstack(icd_df.loc[mask, "icd_embedding"].to_numpy())

    def eval_one_phecode(phecode):
        phe_embedding = phecode_df[phecode_df['Phecode'] == phecode]['phecode_embedding'].to_numpy()[0]
        sims = E @ phe_embedding  # dot == cosine if unit vectors

        icd_sorted = (
            icd_df.loc[mask]
                .assign(cosine_sim=sims)
                .sort_values("cosine_sim", ascending=False)
                .copy()
        )
        icd_sorted['label'] = 0
        icd_sorted.loc[icd_sorted['Phecode'] == phecode, 'label'] = 1

        y_true = np.asarray(icd_sorted['label'], dtype=int)
        scores = np.arange(len(y_true), 0, -1)

        P, R, _ = precision_recall_curve(y_true, scores)
        return R, P, y_true, scores

    # Parallel execution of per‑phecode logic
    results = Parallel(n_jobs=-1, prefer="threads", verbose=1)(
        delayed(eval_one_phecode)(phecode) for phecode in phecodes  # CHANGE
    )

    # Unpack results
    all_curves = []
    y_pool, s_pool = [], []
    for R, P, y_true, scores in results:
        all_curves.append((R, P))
        y_pool.append(y_true)
        s_pool.append(scores)

    # ---------- micro‑average PR curve ------------------------------
    y_all = np.concatenate(y_pool)
    s_all = np.concatenate(s_pool)

    P_micro, R_micro, _ = precision_recall_curve(y_all, s_all)
    ap_micro = average_precision_score(y_all, s_all)

    model_metrics[model_name] = (P_micro, R_micro, ap_micro)

    # ---- Top‑K micro points -------------------------------------------------
    KS = [10, 100, 1000]

    def micro_points_at_ks(y_all, s_all, n_per_query, ks):
        pts = []
        for K in ks:
            if K > n_per_query:
                # skip K that exceeds list length
                continue
            # rank scores are n, n-1, ..., 1  ⇒ Top‑K ⇔ score >= n-K+1
            thr = n_per_query - K + 1
            sel = s_all >= thr
            tp = int(y_all[sel].sum())
            precision = tp / max(int(sel.sum()), 1)
            recall    = tp / max(int(y_all.sum()), 1)
            pts.append((K, recall, precision))
        return pts

    n_per_query = len(s_pool[0])  # all lists have same length
    topk_pts = micro_points_at_ks(y_all, s_all, n_per_query, KS)

    # save both the PR + AP and the Top‑K points
    model_metrics[model_name] = (P_micro, R_micro, ap_micro)
    # add this dict if not already defined above the loop
    try:
        model_topk_points
    except NameError:
        model_topk_points = {}
    model_topk_points[model_name] = topk_pts

```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8, 4))

# sort by AP for legend
sorted_models = sorted(models, key=lambda m: model_metrics[m][2], reverse=True)

# plot curves and remember colors
colors = {}
for model_name in sorted_models:
    P_micro, R_micro, ap_micro = model_metrics[model_name]
    label = rf"{model_name} (AP=$\mathbf{{{ap_micro:.4f}}}$)"
    line, = ax.plot(R_micro, P_micro, label=label, linewidth=1.5)
    colors[model_name] = line.get_color()

# add Top‑K markers (same color as each curve)
for model_name in sorted_models:
    for K, r, p in model_topk_points[model_name]:
        ax.scatter(r, p, color=colors[model_name], s=10, zorder=3)

# (optional) annotate only for the first model to avoid clutter
first = sorted_models[0]
for K, r, p in model_topk_points[first]:
    ax.annotate(f"K={K}", (r, p), xytext=(5, 5),
                textcoords="offset points", fontsize=8)

baseline = y_all.mean()
ax.axhline(baseline, color='black', ls='--', alpha=0.4, label=f'Expected precision (AP=$\mathbf{{{baseline:.4f}}}$)')

ax.set_xlabel("Recall")
ax.set_ylabel("Precision")
ax.set_xlim(0, 1)
ax.set_ylim(-0.05, 1)
ax.grid(True, linestyle="--", alpha=0.5)
ax.legend(frameon=False, loc="best", fontsize=8)
plt.tight_layout()
plt.show()
```

```python
import plotly.graph_objects as go

# ── 1 ▸ a single live figure ──────────────────────────────────────────────
fig = go.FigureWidget(
    px.scatter(
        df, x="UMAP_1", y="UMAP_2",
        hover_name="ICDDescription",
        width=800, height=800,
    )
)
fig.update_traces(marker=dict(size=2.5, opacity=0.8))

# ── 2 ▸ search box ────────────────────────────────────────────────────────
search = widgets.Text(
    description="Find:",
    placeholder="type part of a diagnosis…",
    continuous_update=False,
    layout=widgets.Layout(width="60%")
)

# ── 3 ▸ callback that edits the existing fig ──────────────────────────────
def refresh(change):
    term = change["new"].strip()

    # remove any previous highlight layer(s)
    with fig.batch_update():
        while len(fig.data) > 1:          # keep only the baseline trace
            fig.data = fig.data[:-1]

        if term:
            hits = df[df["ICDDescription"]
                        .str.contains(term, case=False, na=False)]

            if hits.empty:
                print(f"⚠️ No ICDDescription contains “{term}”.")
                return

            # add highlight trace
            fig.add_scattergl(
                x=hits.UMAP_1, y=hits.UMAP_2,
                mode="markers",
                marker=dict(size=2.5, color="red", symbol="diamond"),
                hovertext=hits["ICDDescription"],
                name="matches"
            )

        else:
            # reset axes if search cleared
            fig.layout.xaxis.autorange = True
            fig.layout.yaxis.autorange = True

search.observe(refresh, names="value")

# ── 4 ▸ display UI ────────────────────────────────────────────────────────
display(search, fig)
```
